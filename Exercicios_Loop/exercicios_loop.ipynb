Exercício 12: Elabore um programa que leia quantos números quer que se efetue a soma, subtrações, divisões, multiplicações e no fim por meio de um acumulador diga quantas operações foram efetuadas.
```python
def operacoes():

    num = int(input("Digite um número: "))

    cont = 0

    for i in range(1, num + 1):

        print(num, "+" , i, "=", num + i)
        cont += 1

        print(num, "-" , i, "=", num - i)
        cont += 1

        print(num, "/" , i, "=", num / i)
        cont += 1

        print(num, "*" , i, "=", num * i)
        cont += 1

    print("Total de operacões: ",cont)

operacoes()
```

Exercícios 15: Elabore um programa que escreva no ecrã todas as linhas de código ASCII(0 a 255) e o código correspondente. Dispor de 20 em 20 com a condição de continuação ou saída do programa.

```python
def ascii_table():

    inicio = 0

    while inicio <= 255:

        for i in range(inicio, min(inicio + 20, 256)):
            print(i, "→", chr(i))

        op = input("Continuar? (s/n): ")

        if op == "n":
            break

        inicio += 20


ascii_table()
```

Exercícios 16: Elabore um programa que constitua a média de 30 números pares que sejam introduzidos. Validando a entrada de números inteiros entre 1 e 50.

```python
def media30():

    soma = 0

    for i in range(1, 31):

        num = int(input(f"Digite o {i}º número inteiro: "))

        while num % 2 != 0 or num < 1 or num > 50:
            print("Número inválido!! Deve ser par e estar entre 1 e 50.")
            num = int(input("Digite novamente: "))

        soma += num

    media = soma / 30

    print("Média dos 30 números pares introduzidos:", media)


media30()
```

Exercícios 18: Elabore um programa que leia uma entrada e diga quantos números perfeitos existem. Exemplo de numero perfeito em que somando todos os divisores ele da o numero inicial.

6=3+2+1 .

```python
def num_perfeito():

    num = int(input("Digite um número: "))
    cont = 0

    for i in range(1, num + 1):

        soma = 0

        for j in range(1, i):
            if i % j == 0:
                soma += j

        if soma == i:
            print("Número perfeito:", i)
            cont += 1

    print("Quantidade de números perfeitos:", cont)


num_perfeito()
```

Teste Final: Elabore um programa que leia um valor de entrada e mostre para cada valor até ao 1 (se é número Primo, Quantos divisores e números perfeitos) o Programa deve validar entradas entre 1 e 30.000, e parar de 10 em 10 valores com instrução para parar ou continuar. No mesmo programa use um menu e Elabore uma calculadora simples (+,-,*,/) com a função extra tabuada. Validar entradas de 1 a 1000 (nota a tabuada deve apresentar todas as multiplicações de 1 ate ao máximo introduzido) deve parar de 20 em 20 valores.

Teste Final: Elabore uma base de dados de clientes de uma fábrica de materiais. O programa deverá possibilitar inserção e listagem dos clientes bem como as compras por eles efetuadas( Númcli(Automático), NomCli, morada, tel, nif, compra, Divfin ). Divida final=compra – desconto, valor do desconto se compra for entre 100 e 200 é 5%, se for superior a 200 e inferior a 500 é 10% se superior a 500 é 15%. O programa deve validar todas as entradas e na listagem deve parar cliente a cliente e ser possível busca direta por número de cliente.

```python
def eh_primo(n):
    if n < 2:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True


def contar_divisores(n):
    cont = 0

    for i in range(1, n + 1):
        if n % i == 0:
            cont += 1

    return cont


def eh_perfeito(n):
    soma = 0

    for i in range(1, n):
        if n % i == 0:
            soma += i

    return soma == n


def analisar_numeros():
    num = int(input("Digite um valor entre 1 e 30000: "))

    while num < 1 or num > 30000:
        num = int(input("Valor inválido! Digite novamente: "))

    cont = 0

    for i in range(num, 0, -1):
        print("\nNúmero:", i)
        print("Primo:", "Sim" if eh_primo(i) else "Não")
        print("Quantidade de divisores:", contar_divisores(i))
        print("Perfeito:", "Sim" if eh_perfeito(i) else "Não")

        cont += 1

        if cont % 10 == 0 and i != 1:
            op = input("\nContinuar? (s/n): ").lower()
            if op == "n":
                break


def soma(a, b):
    return a + b


def subtracao(a, b):
    return a - b


def multiplicacao(a, b):
    return a * b


def divisao(a, b):
    if b == 0:
        return "Erro: divisão por zero!"
    return a / b


def tabuada():
    maximo = int(input("Digite o valor máximo da tabuada (1 a 1000): "))

    while maximo < 1 or maximo > 1000:
        maximo = int(input("Valor inválido! Digite novamente: "))

    cont = 0

    for i in range(1, maximo + 1):
        print(f"\nTabuada do {i}")
        for j in range(1, maximo + 1):
            print(f"{i} x {j} = {i * j}")
            cont += 1

            if cont % 20 == 0:
                op = input("\nContinuar? (s/n): ").lower()
                if op == "n":
                    return


def calculadora():
    while True:
        print("\n--- CALCULADORA ---")
        print("1 - Soma")
        print("2 - Subtração")
        print("3 - Multiplicação")
        print("4 - Divisão")
        print("5 - Tabuada")
        print("0 - Voltar")

        op = input("Escolha uma opção: ")

        if op == "0":
            break

        if op == "5":
            tabuada()
            continue

        a = float(input("Digite o 1º valor: "))
        b = float(input("Digite o 2º valor: "))

        if op == "1":
            print("Resultado:", soma(a, b))
        elif op == "2":
            print("Resultado:", subtracao(a, b))
        elif op == "3":
            print("Resultado:", multiplicacao(a, b))
        elif op == "4":
            print("Resultado:", divisao(a, b))
        else:
            print("Opção inválida!")


def menu_principal():
    while True:
        print("\n====== MENU PRINCIPAL ======")
        print("1 - Analisar números")
        print("2 - Calculadora")
        print("0 - Sair")

        op = input("Escolha uma opção: ")

        if op == "1":
            analisar_numeros()
        elif op == "2":
            calculadora()
        elif op == "0":
            print("Programa terminado.")
            break
        else:
            print("Opção inválida!")


menu_principal()
```

```python
clientes = []
num_cli_auto = 1


def calcular_desconto(compra):
    if 100 <= compra <= 200:
        return compra * 0.05
    elif 200 < compra < 500:
        return compra * 0.10
    elif compra >= 500:
        return compra * 0.15
    else:
        return 0


def inserir_cliente():
    global num_cli_auto

    nome = input("Nome do cliente: ").strip()
    while nome == "":
        nome = input("Nome inválido! Digite novamente: ").strip()

    morada = input("Morada: ").strip()
    while morada == "":
        morada = input("Morada inválida! Digite novamente: ").strip()

    tel = input("Telefone: ").strip()
    while not tel.isdigit():
        tel = input("Telefone inválido! Digite novamente: ").strip()

    nif = input("NIF: ").strip()
    while not nif.isdigit() or len(nif) != 9:
        nif = input("NIF inválido! Digite novamente (9 dígitos): ").strip()

    compra = float(input("Valor da compra: "))
    while compra < 0:
        compra = float(input("Valor inválido! Digite novamente: "))

    desconto = calcular_desconto(compra)
    divfin = compra - desconto

    cliente = {
        "numcli": num_cli_auto,
        "nome": nome,
        "morada": morada,
        "tel": tel,
        "nif": nif,
        "compra": compra,
        "desconto": desconto,
        "divfin": divfin
    }

    clientes.append(cliente)
    print(f"Cliente inserido com sucesso! Número de cliente: {num_cli_auto}")

    num_cli_auto += 1


def listar_clientes():
    if len(clientes) == 0:
        print("Não existem clientes registados.")
        return

    for cliente in clientes:
        print("\n----------------------")
        print("Número Cliente:", cliente["numcli"])
        print("Nome:", cliente["nome"])
        print("Morada:", cliente["morada"])
        print("Telefone:", cliente["tel"])
        print("NIF:", cliente["nif"])
        print("Compra:", cliente["compra"])
        print("Desconto:", cliente["desconto"])
        print("Dívida Final:", cliente["divfin"])

        op = input("\nContinuar para o próximo cliente? (s/n): ").lower()
        if op == "n":
            break


def buscar_cliente():
    if len(clientes) == 0:
        print("Não existem clientes registados.")
        return

    numero = input("Digite o número do cliente: ").strip()
    while not numero.isdigit():
        numero = input("Número inválido! Digite novamente: ").strip()

    numero = int(numero)

    for cliente in clientes:
        if cliente["numcli"] == numero:
            print("\nCliente encontrado:")
            print("Número Cliente:", cliente["numcli"])
            print("Nome:", cliente["nome"])
            print("Morada:", cliente["morada"])
            print("Telefone:", cliente["tel"])
            print("NIF:", cliente["nif"])
            print("Compra:", cliente["compra"])
            print("Desconto:", cliente["desconto"])
            print("Dívida Final:", cliente["divfin"])
            return

    print("Cliente não encontrado.")


def menu_clientes():
    while True:
        print("\n====== BASE DE DADOS CLIENTES ======")
        print("1 - Inserir cliente")
        print("2 - Listar clientes")
        print("3 - Buscar cliente por número")
        print("0 - Sair")

        op = input("Escolha uma opção: ")

        if op == "1":
            inserir_cliente()
        elif op == "2":
            listar_clientes()
        elif op == "3":
            buscar_cliente()
        elif op == "0":
            print("Programa terminado.")
            break
        else:
            print("Opção inválida!")


menu_clientes()
```
